```python
csv["city-mpg"] = 0.425143707*csv["city-mpg"]#1 US MPG (US Miles per gallon) is equal to 0.425143707 KMPL (Kilometer per litre)
csv["highway-mpg"] = 0.425143707*csv["highway-mpg"]#1 US MPG (US Miles per gallon) is equal to 0.425143707 KMPL (Kilometer per litre)
```

```python
csv.rename(columns={"city-mpg": "city-kmpl"}, inplace=True)
csv.rename(columns={"highway-mpg": "highway-kmpl"}, inplace=True)
```

```python
bins = np.linspace(csv["price"].min(), csv["price"].max(), 4)
```

```python
bins
```

```python
group_names = ["low", "medium", "high"]
```

```python
csv2 = pd.cut(csv["price"], bins, labels=group_names, include_lowest = True)
```

```python
csv2.to_csv("binned data.csv")
```
